```python
print("Categories in 'sex' variable:  ",end=" ")
print(df['sex'].unique())

print("Categories in 'famsize' variable:  ",end=" ")
print(df['famsize'].unique())

print("Categories in 'Mjob' variable:  ",end=" ")
print(df['Mjob'].unique())

print("Categories in 'age' variable:  ",end=" ")
print(df['age'].unique())


print("Categories in 'freetime' variable:  ",end=" ")
print(df['freetime'].unique())
```

```python
numeric_features = [feature for feature in df.columns if df[feature].dtype !='O' ]
categorical_features = [feature for feature in df.columns if df[feature].dtype =='O' ]

#print columns
print('We have {} numerical features :'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features :'.format(len(categorical_features), categorical_features))
```

```python
df.head(2)
```

```python
df['total score'] =  df['age'] + df['health'] + df['freetime']
df['average'] = df['total score']/3
df.head()
```

```python
good_health = df[df['health'] == 100]['average'].count()
free = df[df['freetime'] == 100]['average'].count()

print(f'Number of student with good health:  {good_health}')
print(f'Number of student with free time:  {free}')
```

Histogram & KDE

```python
fig, axs = plt.subplots(1,2, figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='average',bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='average',bins=30,kde=True,hue='sex')
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='total score',bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='total score',bins=30,kde=True,hue='sex')
plt.show()
```

Female students tend to perform well than male students

```python
fig, axs = plt.subplots(1,3, figsize=(15,7))
plt.subplot(141)
sns.histplot(data=df,x='average',bins=30,kde=True,hue='freetime')
plt.subplot(142)
sns.histplot(data=df[df.sex=='F'],x='average',kde=True,hue='freetime')
plt.subplot(142)
sns.histplot(data=df[df.sex=='M'],x='average',kde=True,hue='freetime')

plt.show()
```

```python
fig, axs = plt.subplots(1,3, figsize=(25,6))
plt.subplot(141)
sns.histplot(data=df,x='average',bins=30,kde=True,hue='health')
plt.subplot(142)
sns.histplot(data=df[df.sex=='F'],x='average',kde=True,hue='health')
plt.subplot(142)
sns.histplot(data=df[df.sex=='M'],x='average',kde=True,hue='health')

plt.show()
```
